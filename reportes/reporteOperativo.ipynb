# Análisis de Datos y Modelado Predictivo del Precio de Viviendas en California

## 1. Introducción

El presente análisis tiene como objetivo explorar y modelar el comportamiento del **precio de las viviendas en California** utilizando técnicas de análisis exploratorio de datos (EDA) y modelos de Machine Learning.

Se busca identificar los factores que influyen en el valor de las viviendas y desarrollar un modelo capaz de **predecir el precio de una vivienda a partir de sus características.**

---

# 2. Descripción del dataset

Variables principales:

| Variable           | Descripción                          |
| ------------------ | ------------------------------------ |
| longitude          | ubicación oeste-este                 |
| latitude           | ubicación norte-sur                  |
| housing_median_age | antigüedad promedio de las viviendas |
| total_rooms        | número total de habitaciones         |
| total_bedrooms     | número total de dormitorios          |
| population         | población del área                   |
| households         | número de hogares                    |
| median_income      | ingreso medio de la zona             |
| median_house_value | valor medio de la vivienda           |
| ocean_proximity    | cercanía al océano                   |


El dataset contiene aproximadamente 20,000 registros.

---

# 3. EDA (Exploratory Data Analysis)

Se realizó un análisis estadístico y visual para comprender la distribución de las variables y detectar posibles patrones.

Principales hallazgos:

- La variable **median_income** presenta la **mayor correlación positiva con el precio de la vivienda (~0.68).**

    

- Las viviendas cercanas al océano tienden a tener **precios más altos.**

    

- Algunas variables presentan **distribuciones sesgadas**, especialmente:

    - total_rooms
    - population
    - households

    

También se identificaron **valores atípicos (outliers)** en varias variables numéricas.

---

# 4. Limpieza y preparación de datos

Se realizaron las siguientes acciones de preprocesamiento:

**Tratamiento de valores nulos**

- Se identificaron valores faltantes en la variable:

    - bedrooms_per_room

    Los valores faltantes fueron imputados utilizando **la mediana**, debido a la presencia de valores atípicos.


**Tratamiento de outliers**

- Se utilizaron métodos basados en **IQR (Rango Intercuartílico)** para detectar valores extremos en variables numéricas.



---

# 5. Feature engineering

Se crearon nuevas variables para mejorar la capacidad explicativa del modelo:

| Variable                 | Descripción                                       |
| ------------------------ | ------------------------------------------------- |
| rooms_per_household      | promedio de habitaciones por hogar                |
| population_per_household | densidad promedio de personas por hogar           |
| bedrooms_per_room        | proporción de dormitorios respecto a habitaciones |


Estas variables permiten capturar relaciones estructurales dentro de los datos que no son evidentes en las variables originales.

---


# 6. Modelos predictivos

Se entrenaron dos modelos de Machine Learning para predecir el precio de las viviendas.

**Modelo 1: Regresión Lineal**

La regresión lineal fue utilizada como modelo base para establecer una referencia de desempeño.

Resultados:
- RMSE: 72,668
- MAE: 50,888
- R²: 0.597

El modelo logra explicar aproximadamente **59% de la variabilidad del precio de las viviendas.**

---
**Modelo 2: Random Forest Regressor**

Se implementó un modelo de **Random Forest**, el cual es capaz de capturar relaciones no lineales entre variables.

Resultados:
- RMSE: 50,382
- MAE: 32,358
- R²: 0.806

Este modelo logra explicar aproximadamente **80% de la variabilidad del precio de las viviendas**, mostrando un desempeño significativamente superior al modelo lineal.



---

# 7. Evaluación de modelos

| Modelo           | RMSE   | MAE    | R²    |
| ---------------- | ------ | ------ | ----- |
| Regresión Lineal | 72,668 | 50,888 | 0.597 |
| Random Forest    | 50,382 | 32,358 | 0.806 |

El modelo **Random Forest** presenta una mejora considerable en todas las métricas evaluadas.

---

# 8. Conclusiones técnicas

- El **ingreso medio del área (median_income)** es el principal factor asociado al precio de las viviendas en California.

- La ingeniería de características permitió mejorar la representación de la información del dataset.

- El modelo **Random Forest** mostró un desempeño superior al modelo de regresión lineal, alcanzando un coeficiente de determinación (R²) de aproximadamente 0.80 para la predicción del precio de las viviendas.

- Estos resultados demuestran que el modelo desarrollado es capaz de capturar relaciones complejas entre las variables y puede utilizarse como una **herramienta predictiva dentro del mercado inmobiliario**.
